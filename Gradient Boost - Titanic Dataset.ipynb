```python
df = pd.concat([df,dummy_sex,Embarked_replace,b],axis=1)
```

```python
df = df.drop(['PassengerId','Name','Sex','Ticket','Cabin_replaced','Embarked'],axis=1)
```

```python
df
```

```python
x=df.drop(['Survived'],axis=1)
y=df['Survived']
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
x_changed = scalar.fit_transform(x)
```

```python
from sklearn.model_selection  import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_changed,y,test_size=0.33,random_state=42)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.ensemble import GradientBoostingClassifier
learning_rates = [0.05,0.1,0.02,0.001,0.01,0.25]
for i in learning_rates:
    gra = GradientBoostingClassifier(random_state=100)
    gra.fit(x_train,y_train)
    print("Learning rate: ",i)
    print("Accuracy score (training) : {0:.3f}".format(gra.score(x_train,y_train)))
    print("Accuracy score (validation) : {0:.3f}".format(gra.score(x_test,y_test)))
    print()
```

At learning 0.5 model get well trained

```python
y_test_pred = gra.predict(x_test)
accuracy_score(y_test,y_test_pred)
```

```python
y_train_pred = gra.predict(x_train)
```

```python
accuracy_score(y_train,y_train_pred)
```
