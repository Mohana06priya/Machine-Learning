```python
from sklearn.cluster import KMeans
from sklearn import metrics
from scipy.spatial.distance import cdist
import numpy as np
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
x1 = np.array ([3,1,1,2,1,6,6,6,5,6,7,8,9,8,9,9,8])
x2 = np.array ([5,4,5,6,5,8,6,7,6,7,1,2,1,2,3,2,3])
x1,x2
```

```python
plt.plot()
plt.title('Dataset')
plt.scatter(x1,x2)
```

```python
x = np.array(list(zip(x1,x2))).reshape(len(x1),2)
x
```

```python
colours = ['b','g','r']
markers = ['o','v','s']
```

```python
distortions = []
K = range(1,10)
for k in K:
    Kmeansmodel = KMeans(n_clusters=k).fit(x)
    distortions.append(sum(np.min(cdist(x,Kmeansmodel.cluster_centers_,'euclidean'),axis=1))/x.shape[0])
```

```python
distortions
```

```python
plt.plot(K,distortions,'bx-')
plt.xlabel('k')
plt.ylabel('Distortions')
plt.title('The Elbow method showing the optimal k')
plt.show()
```

```python
model1 = KMeans(n_clusters=3)
model1.fit(x)
```

```python
centroid = model1.cluster_centers_
centroid
```

```python
label = model1.labels_
label
```

```python
plt.scatter(x1,x2,c=label)
plt.scatter(centroid[:,0],centroid[:,1],c='k')
```
